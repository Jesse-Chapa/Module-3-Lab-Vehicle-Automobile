This program collects user input for an automobile and stores it using a Vehicle superclass
and an Automobile subclass. It then displays the data in a readable format.


```python
class Vehicle:
    def __init__(self, vehicle_type: str):
        self.vehicle_type = vehicle_type


class Automobile(Vehicle):
    def __init__(self, year: str, make: str, model: str, doors: str, roof: str):
        super().__init__("car")  # store "car" in the Vehicle superclass attribute
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof


def get_valid_doors():
    """Ask until the user enters 2 or 4."""
    while True:
        doors = input("Enter number of doors (2 or 4): ").strip()
        if doors in ("2", "4"):
            return doors
        print("Invalid input. Please enter 2 or 4.")


def get_valid_roof():
    """Ask until the user enters solid or sun roof."""
    while True:
        roof = input("Enter roof type (solid or sun roof): ").strip().lower()
        if roof in ("solid", "sun roof", "sunroof"):
            # normalize wording for consistent output
            return "sun roof" if roof in ("sun roof", "sunroof") else "solid"
        print("Invalid input. Please type 'solid' or 'sun roof'.")


def main():
    print("Automobile Entry\n")

    year = input("Enter year: ").strip()
    make = input("Enter make: ").strip()
    model = input("Enter model: ").strip()
    doors = get_valid_doors()
    roof = get_valid_roof()

    car = Automobile(year, make, model, doors, roof)

    print("\nVehicle Information:")
    print(f"  Vehicle type: {car.vehicle_type}")
    print(f"  Year: {car.year}")
    print(f"  Make: {car.make}")
    print(f"  Model: {car.model}")
    print(f"  Number of doors: {car.doors}")
    print(f"  Type of roof: {car.roof}")


main()
```

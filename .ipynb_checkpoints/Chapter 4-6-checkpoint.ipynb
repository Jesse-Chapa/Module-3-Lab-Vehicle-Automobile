## 4.1
Choose a number between 1 and 10 for `secret`. Choose another number between 1 and 10 for `guess`.
Use `if/elif/else` to print:
- "too low" if guess < secret
- "too high" if guess > secret
- "just right" if guess == secret


```python
secret = 7
guess = 4

if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
```

## 4.2
Assign True/False to `small` and `green`.
Use if/else statements to print which one matches:
- cherry
- pea
- watermelon
- pumpkin


```python
small = True
green = False

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif not small and green:
    print("watermelon")
else:
    print("pumpkin")
```

## 6.2
Set `guess_me = 7` and `number = 1`.
Write a `while` loop that compares `number` with `guess_me`:
- print "too low" if number < guess_me
- if number == guess_me: print "found it!" and exit loop
- if number > guess_me: print "oops" and exit loop
Increment `number` at the end of the loop.


```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break

    number += 1
```

## 6.3
Set `guess_me = 5`.
Use a `for` loop to iterate `number` over `range(10)`:
- print "too low" if number < guess_me
- if number == guess_me: print "found it!" and break
- if number > guess_me: print "oops" and exit the loop


```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print("too low")
    elif number == guess_me:
        print("found it!")
        break
    else:
        print("oops")
        break
```

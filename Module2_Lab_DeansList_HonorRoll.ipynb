# Module 2 Lab – Case Study: if...else and while
This program accepts student names and GPAs and determines whether
a student qualifies for the Dean’s List or Honor Roll.


```python
# Name: Jesse Chapa
# File Name: module2_lab_gpa.py
# Description: This program accepts student names and GPAs and determines
#  whether the student qualifies for the Dean's List or Honor Roll.

while True:
    last_name = input("Enter student's last name (or ZZZ to quit): ")

    if last_name == "ZZZ":
        print("Exiting program.")
        break

    first_name = input("Enter student's first name: ")

    gpa = float(input("Enter student's GPA: "))

    if gpa >= 3.5:
        print(first_name, last_name, "has made the Dean's List.")
    elif gpa >= 3.25:
        print(first_name, last_name, "has made the Honor Roll.")
    else:
        print(first_name, last_name, "did not qualify for honors.")

    print()  # blank line for readability
```
